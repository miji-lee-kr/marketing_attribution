### Introduction

The dataset is from an online beauty store in Germany, which includes customer journey from various marketing channels, and each customers' total purchase amount. 
The purpose of the analysis is to find out the most effective channel, in order to maximise the impact of our marketing cost spending.

### Data Preprocessing

```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
df = pd.read_csv('attribution.csv')
df.info()
```

```python
df.drop_duplicates(inplace = True)
df[df.isnull()].count()
```

```python
# handling time & date

from datetime import datetime, timedelta

df['time'] = pd.to_datetime(df['time'])

df['dayofweek'] = df['time'].dt.dayofweek
df['hour'] = df['time'].dt.hour

def week_number_of_month(date_value):
     return (date_value.isocalendar()[1] - date_value.replace(day=1).isocalendar()[1])
df['weekofmonth'] = df['time'].apply(week_number_of_month)
```

```python
def week_eng(date_value):
    if date_value == 0:
        return "0_Mon"
    elif date_value == 1:
        return "1_Tue"
    elif date_value == 2:
        return "2_Wed"
    elif date_value == 3:
        return "3_Thu"
    elif date_value == 4:
        return "4_Fri"
    elif date_value == 5:
        return "5_Sat"
    else:
        return "6_Sun"

df['dayofweek'] = df['dayofweek'].apply(week_eng)
```

```python
bins = [0,7,12,18,22,24]
labels = ['5_night', '1_morning', '2_afternoon','3_evening', '4_eve-night']
df['hour_group'] = pd.cut(df['hour'], bins, right = False, labels = labels)

df.head()
```

### Conversion Attribution

```python
# step 1. deciding attribution window for each channel.

df['time'] = pd.to_datetime(df['time'])
successful_conversions = df[df['conversion'] == 1]

attribution_windows = {
    'Instagram': timedelta(days=28),
    'Online Display': timedelta(days=28),
    'Paid Search': timedelta(days=28),
    'Facebook': timedelta(days=28),
    'Online Video': timedelta(days=28)}

# I assigned 28 days to all channels, based on Facebook's attribution window value of 28 days for clicks 
```

```python
def assign_attribution_window(row):
    interaction_time = row['time']
    channel = row['channel']
    attribution_window = attribution_windows[channel]
    attribution_start_time = interaction_time + attribution_window
    return attribution_start_time

df['attribution_validity'] = df.apply(assign_attribution_window, axis=1)

conversion_times = df.loc[df['conversion'] == 1, ['cookie', 'time']].groupby('cookie').min()
merged_df = pd.merge(df, conversion_times, on='cookie', suffixes=('', '_conv'))
filtered_rows = merged_df[merged_df['attribution_validity'] >= merged_df['time_conv']]

df['within_window'] = df.cookie.isin(filtered_rows.cookie)
df.head()

# this is to use attribution windows that we calculated earlier, for single & multi touch
```

```python
# step 2. Last Click Model 
# For non-data-driven attribution model, I'm going to use Last click model, which is most commonly used in many tracking tools

conversion_df = df[df['interaction'] == 'conversion']
last_click_df = conversion_df.groupby('cookie').last()
last_click_result = last_click_df['channel'].value_counts().reset_index()
last_click_result.columns = ['Channel', 'Conversions']
total_conversions = last_click_result['Conversions'].sum()

last_click_result['Attribution'] = last_click_result['Conversions'] / total_conversions
last_click_result
```

```python
last_click_result = last_click_result.sort_values(by='Attribution', ascending=False)

# Visualize the results with a bar chart
sns.barplot(x='Channel', y='Attribution', data=last_click_result, color = 'blue')
plt.xlabel('Channels')
plt.ylabel('Attribution')
plt.title('Last Click Revenue Attribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# step 3. Data Driven Attribution Model 

# logistic regression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

train_df = train_df.dropna()
encoder = LabelEncoder()
train_df['channel_encoded'] = encoder.fit_transform(train_df['channel'])

feature = train_df['channel_encoded']
output = train_df['conversion']
model = sm.formula.glm("output ~ feature", train_df, family=sm.families.Binomial()).fit()
print(model.summary())

# since p-value of feature is greater than 0.05, the result is not statistically significant 
# probably because it doesn't take into consideration that users can convert after visiting from the same channel multiple times 
# I have to try Markov chain analysis
```

```python
# step 4. Markov Chain analysis

# Get a transition matrix which reflects the day of week impact

df = df.sort_values(['cookie', 'time'])

df['next_channel'] = df.groupby('cookie')['channel'].shift(-1)
df['day_of_week'] = df['time'].dt.dayofweek

channels = df['channel'].unique()
days_of_week = df['day_of_week'].unique()

transition_matrices = {}

for day in days_of_week:
    df_day = df[df['day_of_week'] == day]
    transition_matrix = pd.DataFrame(index=channels, columns=channels, dtype=float)
    
    for channel in channels:
        channel_counts = df_day[df_day['channel'] == channel]['next_channel'].value_counts()
        transition_matrix.loc[channel] = channel_counts
        
    transition_matrix = transition_matrix.fillna(0)
    transition_matrix = transition_matrix.div(transition_matrix.sum(axis=1), axis=0)
    
    transition_matrices[day] = transition_matrix
```

```python
attribution_adjustments = {}

for day, transition_matrix in transition_matrices.items():
    attribution_weight_adjustment = pd.DataFrame(np.ones_like(transition_matrix.values),
                                                index=transition_matrix.index,
                                                columns=transition_matrix.columns)
    
    for channel, _ in transition_matrix.iterrows():
        attribution_weight_adjustment.loc[channel] = 0
        remaining_weights_sum = attribution_weight_adjustment.sum(axis=1).values
        attribution_weight_adjustment.loc[channel] = 1 - remaining_weights_sum
    attribution_weight_adjustment = attribution_weight_adjustment.div(attribution_weight_adjustment.sum(axis=1),
                                                                     axis=0)
    
    attribution_adjustments[day] = attribution_weight_adjustment
```

```python
# Apply attribution adjustments to conversion rate

df['conversion_rate_adjusted'] = df['conversion_value'] / df.groupby('channel')['conversion_value'].transform('sum')

for day, attribution_adjustment in attribution_adjustments.items():
    mask = df['day_of_week'] == day
    channel_column = df.loc[mask, 'channel']
    adjustment_values = attribution_adjustment.lookup(channel_column, channel_column)
    df.loc[mask, 'conversion_rate_adjusted'] *= adjustment_values

adjusted_channel_conversion_rate = df.groupby('channel')['conversion_rate_adjusted'].mean()
adjusted_channel_conversion_rate.plot(kind='bar');
```

```python
total_conversion_rate = adjusted_channel_conversion_rate.sum()
attribution_conversion_rate = (adjusted_channel_conversion_rate / total_conversion_rate) * 100

attribution_conversion_rate
```

```python
# Create the 'attribution_value_adjusted' column with the original conversion values
df['attribution_value_adjusted'] = df['conversion_value']

# Apply attribution adjustments to conversion values
for day, adjusted_transition_matrix in attribution_adjustments.items():
    mask = df['day_of_week'] == day
    channel_column = df.loc[mask, 'channel'].astype(str)  # Convert channel labels to strings
    next_channel_column = df.loc[mask, 'next_channel'].astype(str)  # Convert next channel labels to strings
    attribution_value_column = df.loc[mask, 'attribution_value_adjusted']
    
    # Convert the channel labels to indices to perform the lookup
    channel_indices = np.searchsorted(adjusted_transition_matrix.index.astype(str), channel_column.astype(str))
    next_channel_indices = np.searchsorted(adjusted_transition_matrix.columns.astype(str), next_channel_column.astype(str))
    
    # Filter out of range indices
    valid_indices = (channel_indices < len(adjusted_transition_matrix.index)) & (next_channel_indices < len(adjusted_transition_matrix.columns))
    channel_indices = channel_indices[valid_indices]
    next_channel_indices = next_channel_indices[valid_indices]
    attribution_value_column = attribution_value_column[valid_indices]
    
    # Perform the lookup using the indices
    adjusted_values = adjusted_transition_matrix.values[channel_indices, next_channel_indices]
    
    df.loc[mask, 'attribution_value_adjusted'] = attribution_value_column * adjusted_values

# Calculate the adjusted conversion value for each channel
adjusted_channel_values = df.groupby('channel')['attribution_value_adjusted'].sum()

# Normalize the adjusted_channel_values to sum up to 100%
normalized_adjusted_channel_values = adjusted_channel_values / adjusted_channel_values.sum() * 100

# Display the adjusted conversion value for each channel
print("Adjusted Conversion Value by Channel:")
print(normalized_adjusted_channel_values)

```

```python
# Markov Chain Analysis

df['next_channel'] = df.groupby('cookie')['channel'].shift(-1)
channels = df['channel'].unique()
transition_matrix = pd.DataFrame(index=channels, columns=channels, dtype=float)

for channel in channels:
    channel_counts = df[df['channel'] == channel]['next_channel'].value_counts()
    transition_matrix.loc[channel] = channel_counts
    
transition_matrix = transition_matrix.fillna(0)
transition_matrix = transition_matrix.div(transition_matrix.sum(axis=1), axis=0)

conversion_values = df.groupby('channel')['conversion_value'].sum()
transition_matrix = transition_matrix.values
conversion_values = conversion_values.values
num_channels = transition_matrix.shape[0]
initial_probabilities = np.ones(num_channels) / num_channels

stationary_distribution = initial_probabilities
for _ in range(100):  
    stationary_distribution = np.dot(stationary_distribution, transition_matrix)
stationary_distribution /= np.sum(stationary_distribution)

revenue_attributions = stationary_distribution * conversion_values
attribution_table = pd.DataFrame({'Channel': channels, 'Revenue Attribution': revenue_attributions})

attribution_table = attribution_table.sort_values(by='Revenue Attribution', ascending=False)
attribution_table
```
